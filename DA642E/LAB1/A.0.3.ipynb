```python
import random as rand
def flipCoin():
  hCounter = 0
  tCounter = 0
  flipList = []

  while hCounter < 4 or tCounter < 4:
    randomInteger = rand.randint(0,1)
    if randomInteger == 0:
      hCounter+=1;
      flipList.append('H')
    else:
      tCounter+=1;
      flipList.append('T')

  print(f"Result: {flipList}\nNumber of flips: {len(flipList)}\n")

for i in range(10):
  flipCoin()
```
