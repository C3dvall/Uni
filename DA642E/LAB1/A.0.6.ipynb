```python
import random as rand
def createDeck():
  values = ['2', '3', '4', '5', '6', '7', '8', '9', 'T', 'J', 'Q', 'K', 'A']
  suits = ['s', 'h', 'd', 'c']
  deck = []

  for suit in suits:
      for value in values:
          card = value + suit
          deck.append(card)
  print(deck)
  shuffler(deck)
  print(deck)

def shuffler(deck):
  length = len(deck)
  for i in range(0, length):
    r = rand.randint(0, length - 1)
    temp = deck[i]
    deck[i] = deck[r]
    deck[r] = temp
  return deck

createDeck()
```
