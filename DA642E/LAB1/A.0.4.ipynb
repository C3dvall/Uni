```python
isPrime = 0

def isPrimeNumber(isPrime):
  if isPrime > 1:
    is_prime = True
    for i in range(2, int(isPrime**0.5) + 1):
      if (isPrime % i) == 0:
        is_prime = False
        break
    if is_prime:
      print(isPrime, "is a prime number")
      return True
    else:
      print(isPrime, "is not a prime number")
      return False
  else:
    print(isPrime, "is not a prime number")
    return False


while(isPrime > -1):
  isPrime =int(input("Enter an integer: "))
  if isPrime != -1:
    isPrimeNumber(isPrime)
  else:
    print("Exiting")
    break
```
