```python
import numpy as np

cinnamonBun = 35
discount = 0.6

def getAmountOfBuns():
  nbrOfBuns = float(input("How many Cinnamon buns do you want? "))
  if(nbrOfBuns != int(nbrOfBuns) or nbrOfBuns == 0):
    print("You can only buy whole buns")
    return 0
  else:
    return nbrOfBuns

def calculateTotalprice():
  nbrOfBuns = float = getAmountOfBuns()
  if(nbrOfBuns == 0):
    print("You did not buy any buns")
  else:
    total = np.multiply(cinnamonBun, nbrOfBuns)
    print("The total price if you had bought fresh Cinnamon buns would have been: %.2f" % total)
    print("The price for buying a day old Cinnamon buns: %.2f" % (total*discount))

calculateTotalprice()
```
