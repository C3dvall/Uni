```python
import numpy as np

firstTwoYearsMultiplier = 10.5
additionalYearsMultiplier = 4

def dogYearCalculator():
  years = float(input("How many years old is your dog? "))
  if years <= 2 and years >=1:
    return print(f"Your dog is: {np.multiply(years, firstTwoYearsMultiplier):.2f} years old!")
  elif years <= 0:
    return print("Please enter a positive number!")
  else:
    firstYears = 2 * firstTwoYearsMultiplier
    additionalYears = (years - 2) * additionalYearsMultiplier
    return print(f"Your dog is: {np.add(firstYears, additionalYears):.2f} years old!")


dogYearCalculator()
```
